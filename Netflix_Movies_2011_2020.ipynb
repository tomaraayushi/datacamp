```python
# Create a list of years from 2011-2020
years = []
year = 2011
for i in range(10):
  years.append(year)
  year += 1
years
```

```python
# Create a list of average movie length 'durations'
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
```

```python
# Create a dictionary 'movie_dict'
movie_dict = {}
movie_dict['years'] = years
movie_dict['durations'] = durations
movie_dict
```

```python
# Import pandas to create dataframe using dictionary'movie_dict'
import pandas as pd

durations_df = pd.DataFrame(movie_dict)
durations_df
```

```python
# Import matplotlib to visualize the data
import matplotlib.pyplot as plt

# Draw a line plot of release years and dureations

plt.plot(durations_df['years'], durations_df['durations'])

# Add a title tp the figure
plt.title("Netflix Movie Durations 2011-2020")
plt.xlabel('release_year')
plt.ylabel('durations')
plt.show()

```

```python
# Load more data from csv file
# Create new dataframe 'netfilx_df'

netflix_df = pd.read_csv('netflix_data.csv')

# Get first five rows to get an idea of the data
netflix_df.head()
```

```python
#Subset the netflix_df to get the movies only

netflix_df_movies = netflix_df.loc[netflix_df['type'] == 'Movie']

# Select columns in netflix_df_movies dataframe

netflix_movies_col = netflix_df_movies[['title', 'country', 'genre', 'release_year', 'duration']]

netflix_movies_col.head()
```

```python
#Visualize movie durations w.r.t release year

fig = plt.scatter(netflix_movies_col['release_year'], netflix_movies_col['duration'])

plt.title("Movie Duration by Year of Release")
plt.xlabel('release_year')
plt.ylabel('durations')

plt.show()
```

```python
#Subset netflix_movies_col to create a new df for short movies because as we can see in the scatter plot that more short movies are released in past decades

short_movies = netflix_movies_col[netflix_movies_col['duration'] < 60]

short_movies.head(20)
```

```python
# Create a new list
colours = []

# Iterate through the netflix_movies_col to find the movies in each genre

for idx, gnr in netflix_movies_col.iterrows():
  if gnr['genre'] == 'Children':
    colours.append('red')
  elif gnr['genre'] == 'Documentaries':
    colours.append('blue')
  elif gnr['genre'] == 'Stand-Up':
    colours.append('green')
  else:
    colours.append('black')

colours[0:10]
```

```python
#Visualize it again with fewer modifications

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize= (12,8))

plt.scatter(netflix_movies_col['release_year'], netflix_movies_col['duration'], color= colours)

#Create title and add axis-labels

plt.title("Movie Duration by Year of Release")
plt.xlabel("Release Year")
plt.ylabel("Duration(min)")

#Show the scatter plot

plt.show()
```
